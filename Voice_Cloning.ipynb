```python
from TTS.api import TTS
```

```python
tts = TTS("tts_models/multilingual/multi-dataset/bark", gpu=True)
```

```python
text = "Hello my name is Obama and i am helping Saurabh to get this internship"
```

```python
tts.tts_to_file(text=text,                                     # Text to Clone
                file_path="output.wav",                        # Output file name
                voice_dir="/content/bark_voices",              # Speaker file directory  bark_voices/new_speaker/speaker.wav
                speaker="new_speaker",                         # Name of the speaker directory
                speed = 1.0,                                   # Speed of the desired audio
                emotion = "Happy")                           # Emotion
```
